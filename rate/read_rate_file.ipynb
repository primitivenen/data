```python
import pandas as pd
```

```python
data = pd.read_csv("mid2 max rate.csv", header=None)
```

```python
data.head()
```

```python
data.columns=['raw']
```

```python
data.head()
```

```python
data[0:1]
```

```python
def f2(row):
    rate = float(row['raw'].split("is")[1])
    return rate
```

```python
def f(row):
    vin = row['raw'].split(",")[0].split(":")[1]
    return vin
```

```python
data['rate'] = data.apply(f2, axis=1)
```

```python
data['vin'] = data.apply(f, axis=1)
```

```python
data = data.drop(data.columns[0], axis=1)
```

```python
data.sort_values(by='rate', ascending=False)
```
