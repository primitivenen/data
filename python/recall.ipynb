```python
import numpy as np  
import pandas as pd  
import matplotlib.pyplot as plt  
import seaborn as sb  
from scipy.io import loadmat  
%matplotlib notebook
import matplotlib as mpl
mpl.get_backend()

data = pd.read_csv("C:\Users\Wan Chen\Desktop\project\Recall-data-2018.1.23.csv")
data.shape 
```

```python
data.head()
```

```python
data = data.assign(CoarseCategory=data['DefectiveParts'].str.split(':').str.get(0))
```

```python
data.head()
```

```python
data = data.assign(FineCategory=data['DefectiveParts'].str.split(':').str.get(1))
#data = data.assign(FineCategory=data['DefectiveParts'].str.slice(str.find(':') + 1))
```

```python
data.head()
```

```python
fig=plt.figure()
data['CoarseCategory'].value_counts().plot(kind='bar')
plt.xlabel('Coarse Category')
plt.ylabel('Frequency')
plt.title('Distribution of Coarse Categories for Defective Parts')
fig.savefig('test.jpg')
```

```python
plt.figure()
data.groupby('CoarseCategory').size().plot(kind='bar')
```

```python
%matplotlib inline
plt.show()
```

```python
plt.figure()
import seaborn as sns
sns.set(style="darkgrid")
ax=sns.countplot(x="CoarseCategory",data=data, order = data['CoarseCategory'].value_counts().index)
#sns.set(font_scale=0.001)
#plt.xticks(rotation=90)
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right", fontsize=6)
plt.tight_layout()
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure()
import seaborn as sns
sns.set(style="darkgrid")
ax=sns.countplot(y="CoarseCategory",data=data, order = data['CoarseCategory'].value_counts().index)
sns.set(font_scale=0.001)
#plt.xticks(rotation=90)
ax.set_yticklabels(ax.get_yticklabels(), fontsize=7)
plt.tight_layout()
plt.xlabel("Frequency")
plt.show()
```

```python
plt.figure()
import seaborn as sns
sns.set(style="darkgrid")
ax=sns.countplot(y="FineCategory",data=data[data['CoarseCategory']=='EQUIPMENT'], order = data['FineCategory'].value_counts().index)
ax.set_yticklabels(ax.get_yticklabels(), fontsize=9)
plt.tight_layout()
plt.xlabel("Frequency")
plt.show()
```
